# Data Loading and Preprocessing

```python
train = pd.read_csv("/kaggle/input/playground-series-s6e2/train.csv")
test = pd.read_csv("/kaggle/input/playground-series-s6e2/test.csv")
original = pd.read_csv("/kaggle/input/datasets/wangxintong111/heartdisease/Heart_Disease_Prediction.csv") 

# 1. Label Encoding
le = LabelEncoder()
train['Heart Disease'] = le.fit_transform(train['Heart Disease'])
original['Heart Disease'] = le.fit_transform(original['Heart Disease'])

base_features = [col for col in train.columns if col not in ['Heart Disease', 'id']] 

# 2. æå–ç»Ÿè®¡ç‰¹å¾ (Target Encoding)
def add_engineered_features(df):
    df_temp = df.copy()
    
    for col in base_features: 
        if col in original.columns:
            stats = original.groupby(col)['Heart Disease'].agg(['mean', 'median', 'std', 'skew', 'count']).reset_index()
            stats.columns = [col] + [f"orig_{col}_{s}" for s in ['mean', 'median', 'std', 'skew', 'count']]
            df_temp = df_temp.merge(stats, on=col, how='left') 
 
            fill_values = {
                f"orig_{col}_mean": original['Heart Disease'].mean(),
                f"orig_{col}_median": original['Heart Disease'].median(),
                f"orig_{col}_std": 0,
                f"orig_{col}_skew": 0,
                f"orig_{col}_count": 0
            }
            df_temp = df_temp.fillna(value=fill_values)
            
    return df_temp

train = add_engineered_features(train)
test = add_engineered_features(test) 

# ==========================================
# ðŸŒŸ æ ¸å¿ƒä¿®æ”¹ï¼šå°† original çœŸå®žæ•°æ®å®Œç¾Žæ‹¼æŽ¥å…¥ train
# ==========================================
original_for_concat = original.copy()

# æ‰¾å‡º train ç”Ÿæˆäº†ä½† original æ²¡æœ‰çš„åˆ— (å³å„ç§ orig_xxx ç»Ÿè®¡åˆ—)
missing_cols = [col for col in train.columns if col not in original_for_concat.columns and col != 'id']

# ç»™ original è¡¥é½è¿™äº›åˆ—ï¼Œå¹¶å¡«å……å®‰å…¨é»˜è®¤å€¼ï¼Œé˜²æ­¢å‡ºçŽ° NaN
for col in missing_cols:
    if 'mean' in col:
        original_for_concat[col] = original['Heart Disease'].mean()
    elif 'median' in col:
        original_for_concat[col] = original['Heart Disease'].median()
    else:
        original_for_concat[col] = 0

# åŽ»é™¤åŽŸæ•°æ®çš„ id åˆ—ï¼ˆå¦‚æžœå­˜åœ¨ï¼‰
if 'id' in original_for_concat.columns:
    original_for_concat = original_for_concat.drop(columns=['id'])

# ç›´æŽ¥æ‹¼æŽ¥åŽŸå§‹æ•°æ®ï¼è®©æ¨¡åž‹è§è¯†çœŸå®žçš„æ‚£è€…ï¼
train = pd.concat([train, original_for_concat], axis=0).reset_index(drop=True)
# ==========================================

# 3. ç”Ÿæˆ X å’Œ y
X = train.drop(['id', 'Heart Disease'], axis=1, errors='ignore')
y = train['Heart Disease']
X_test = test.drop(['id'], axis=1, errors='ignore')

print(f"âœ… æ•°æ®å¤„ç†ä¸Žæ‹¼æŽ¥å®Œæˆï¼")
print(f"   å½“å‰è®­ç»ƒé›† X å½¢çŠ¶: {X.shape} (å·²åŒ…å« Original æ ·æœ¬)")
print(f"   å½“å‰æµ‹è¯•é›† X_test å½¢çŠ¶: {X_test.shape}")
```

# Data Quality Check (Missing & Duplicates)

```python
def check_data_quality(df, name="Dataset"):
    print(f"--- Data Quality: {name} ---")
    print(f"Total Rows: {len(df)}")

    cols_to_check = [c for c in df.columns if c != 'id']
    dupes = df.duplicated(subset=cols_to_check).sum()

    nan_counts = df.isnull().sum()
    total_nans = nan_counts.sum()
    
    print(f"Duplicate Rows (excl. ID): {dupes}")
    print(f"Total NaN values: {total_nans}")
    if total_nans > 0:
        print("\nColumns with NaNs:")
        print(nan_counts[nan_counts > 0])
    print("-" * 30)

check_data_quality(train, "Train")
check_data_quality(test, "Test")
```

# Feature Uniqueness & Cardinality

```python
def analyze_uniqueness(df):
    unique_stats = []
    for col in df.columns:
        if col == 'id': continue
        
        n_unique = df[col].nunique()
        dtype = df[col].dtype

        category_guess = "Categorical/Ordinal" if n_unique < 25 else "Continuous"
        
        unique_stats.append({
            'Feature': col,
            'Unique Values': n_unique,
            'Data Type': dtype,
            'Heuristic Type': category_guess
        })
    
    return pd.DataFrame(unique_stats).sort_values(by='Unique Values')

uniqueness_df = analyze_uniqueness(train)
print(uniqueness_df)
```

# Cross-Validation & Training

```python
%%time
import torch
import gc
import numpy as np
import pandas as pd
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import roc_auc_score
from joblib import Parallel, delayed

# 1. çŽ¯å¢ƒæ£€æŸ¥
n_gpus = torch.cuda.device_count()
print(f"ðŸŸ¢ Detected {n_gpus} GPUs. Mode: {'Parallel' if n_gpus > 1 else 'Sequential'}")

# 2. åŸºç¡€å‚æ•°
param_grid = {
    'random_state': 42,
    'verbosity': 1, 
    'n_epochs': 100,
    'batch_size': 256, 
    'n_ens': 8, 
    'use_early_stopping': True,
    'early_stopping_additive_patience': 20,
    'early_stopping_multiplicative_patience': 1,
    'act': "mish",
    'embedding_size': 8,
    'first_layer_lr_factor': 0.5962121993798933,
    'hidden_sizes': "rectangular",
    'hidden_width': 384,
    'lr': 0.04, 
    'ls_eps': 0.011498317194338772,
    'ls_eps_sched': "coslog4",
    'max_one_hot_cat_size': 18,
    'n_hidden_layers': 4, 
    'p_drop': 0.07301419697186451,
    'p_drop_sched': "flat_cos",
    'plr_hidden_1': 16, 
    'plr_hidden_2': 8,
    'plr_lr_factor': 0.1151437622270563,
    'plr_sigma': 2.3316811282666916,
    'scale_lr_factor': 2.244801835541429,
    'sq_mom': 1.0 - 0.011834054955582318,
    'wd': 0.02369230879235962,
} 

# ==========================================
# ðŸŒŸ [INSERTED] ALL 9 NEJM Medical Features
# ==========================================
def add_nejm_features_full(df):
    df_eng = df.copy()
    
    # 1. æ™ºèƒ½åŒ¹é…åˆ—å
    cols = df.columns
    bp_col = [c for c in cols if 'BP' in c or 'Pressure' in c][0]
    chol_col = [c for c in cols if 'Chol' in c][0]
    fbs_col = [c for c in cols if 'FBS' in c or 'Sugar' in c][0]
    sex_col = [c for c in cols if 'Sex' in c or 'gender' in c][0]
    
    # 2. åŸºç¡€äºŒå€¼å› å­ (Binary Flags)
    df_eng['Risk_High_BP'] = (df_eng[bp_col] >= 140).astype(int)
    df_eng['Risk_High_Chol'] = (df_eng[chol_col] >= 240).astype(int)
    
    # å…¼å®¹æ•°å€¼åž‹å’Œåˆ†ç±»åž‹ FBS
    if df_eng[fbs_col].max() > 1:
        df_eng['Risk_Diabetes'] = (df_eng[fbs_col] > 120).astype(int)
    else:
        df_eng['Risk_Diabetes'] = df_eng[fbs_col].astype(int)

    # 3. ç´¯ç§¯é£Žé™©è´Ÿè· (Cumulative Score)
    df_eng['Risk_Burden_Score'] = (
        df_eng['Risk_High_BP'] +
        df_eng['Risk_High_Chol'] +
        df_eng['Risk_Diabetes']
    )

    # 4. S-Tier: å…³é”®å¹´é¾„çª—å£ (55-60)
    df_eng['Age_Window_55_60'] = ((df_eng['Age'] >= 55) & (df_eng['Age'] < 60)).astype(int)
    df_eng['Critical_Age_Hypertension'] = df_eng['Age_Window_55_60'] * df_eng['Risk_High_BP']

    # 5. ä¸­å¹´åŸºçº¿ (Age 50+)
    df_eng['Age_50_Plus'] = (df_eng['Age'] >= 50).astype(int)
    df_eng['Age50_Risk_Interaction'] = df_eng['Age_50_Plus'] * df_eng['Risk_Burden_Score']

    # 6. A-Tier: æ€§åˆ«äº¤äº’ (Sex x Risk)
    df_eng['Sex_Risk_Interaction'] = df_eng[sex_col].astype(int) * df_eng['Risk_Burden_Score']

    return df_eng

# åº”ç”¨ç‰¹å¾å·¥ç¨‹ (åœ¨è½¬ categorical ä¹‹å‰ï¼)
X = add_nejm_features_full(X)
X_test = add_nejm_features_full(X_test)
print("âœ… Added 9 NEJM Medical Features.")
# ==========================================

# 3. æ•°æ®é¢„å¤„ç†
X = X.reset_index(drop=True)
y = y.reset_index(drop=True)

# ðŸš¨ å…³é”®æ­¥éª¤ï¼šå…¨é‡ Label Embedding åŒ–
# è¿™é‡Œçš„å¾ªçŽ¯ä¼šè‡ªåŠ¨æŠŠåˆšæ‰æ·»åŠ çš„ 9 ä¸ªæ–°ç‰¹å¾å…¨éƒ¨è½¬ä¸º category ç±»åž‹
for col in X.columns:
    X[col] = X[col].astype(str).astype('category')
    X_test[col] = X_test[col].astype(str).astype('category') 

skf = StratifiedKFold(n_splits=N_FOLDS, shuffle=True, random_state=RANDOM_STATE)

# 4. å®šä¹‰å•ä¸ª Fold çš„è®­ç»ƒå‡½æ•°
def train_single_fold(fold, train_idx, val_idx):
    device_id = fold % n_gpus
    device = f'cuda:{device_id}'
    
    print(f"ðŸš€ Fold {fold + 1} started on {device}...")
    
    fold_params = param_grid.copy()
    fold_params['device'] = device
    
    X_tr, X_val = X.iloc[train_idx], X.iloc[val_idx]
    y_tr, y_val = y.iloc[train_idx], y.iloc[val_idx]
    
    model = RealMLP_TD_Classifier(**fold_params) 
    model.fit(X_tr, y_tr.values, X_val, y_val.values) 
    
    val_probs = model.predict_proba(X_val)[:, 1]
    test_probs = model.predict_proba(X_test)[:, 1]
    
    score = roc_auc_score(y_val, val_probs)
    print(f"âœ… Fold {fold + 1} finished on {device}. AUC: {score:.5f}")
    
    del model, X_tr, X_val, y_tr, y_val
    gc.collect()
    if device.startswith('cuda'):
        try:
            torch.cuda.set_device(device)
            torch.cuda.empty_cache()
        except:
            pass
            
    return val_idx, val_probs, test_probs, score

# 5. å¹¶è¡Œæ‰§è¡Œ
results = Parallel(n_jobs=n_gpus, backend="loky")(
    delayed(train_single_fold)(fold, train_idx, val_idx) 
    for fold, (train_idx, val_idx) in enumerate(skf.split(X, y))
)

# 6. æ±‡æ€»ç»“æžœ
oof_preds = np.zeros(len(train))
test_preds = np.zeros(len(test))
fold_scores = []

for val_idx, val_probs, fold_test_probs, score in results:
    oof_preds[val_idx] = val_probs
    test_preds += fold_test_probs / N_FOLDS
    fold_scores.append(score)

print(f"\n" + "="*40)
print(f"Mean AUC: {np.mean(fold_scores):.5f}")
print("="*40)
```

```python
from sklearn.calibration import CalibrationDisplay

plt.figure(figsize=(12, 5)) 
plt.suptitle('RealMLP', fontsize=20) 

ax1 = plt.subplot(1, 2, 1)
CalibrationDisplay.from_predictions(y, oof_preds, n_bins=100, strategy='quantile', ax=ax1)
ax1.set_title('Calibration Curve')

ax2 = plt.subplot(1, 2, 2)
ax2.hist(oof_preds, bins=100, edgecolor='black', alpha=0.7)
ax2.set_title('Prediction Distribution (Histogram)')
ax2.set_xlabel('Predicted Probability')
ax2.set_ylabel('Count')

plt.tight_layout(rect=[0, 0.03, 1, 0.95]) 
plt.show()
```
